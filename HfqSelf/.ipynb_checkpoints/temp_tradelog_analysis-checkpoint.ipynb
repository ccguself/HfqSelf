```python
import numpy as np
import pandas as pd
import os
from peewee import *
import datetime
```

```python
db = SqliteDatabase("D:\Trader\.vntrader\database.db")
```

```python
class SignalDb(Model):
    signal_id = CharField(primary_key=True)
    strategy = CharField()  # 产生该信号的策略名称
    symbol = CharField()  # 标的名称（rb2105.SHFE）
    
    
    datetime_begin = DateTimeField()  # 信号作用区间-开始
    datetime_begin = DateTimeField()  # 信号作用区间-结束
    direction = CharField()  # 方向
    price_limit_1 = FloatField()  # 开仓价格
    price_limit_2 = FloatField()  # 平仓价格
    price_stop = FloatField()  # 强制止损价格
    
    volume_signal = IntegerField()  # 信号下单量
    duration_min = IntegerField()  # 持仓时间/分钟
    cancel_second = IntegerField()  # 撤单时间/秒
    
    used_to_trade = BooleanField()  # 是否触发交易
    
    class Meta:
        database = db

class TradeSimulateDb(Model):
    strategy = CharField()  # 产生该信号的策略名称
    symbol = CharField()  # 标的名称（rb2105.SHFE）
    
    datetime_simulate_trade = DateTimeField()  # 模拟盘中，发生该笔交易的时间
    datetime_triger_tick = DateTimeField()  # 模拟盘中，触发该交易的tick的时间

    direction = CharField()  # 多空方向
    offset = CharField()  # 开平仓方向
#     price_trade_ideal = FloatField()  # 理想成交价格
    price_trade_simulate = FloatField()  # 模拟盘成交价格
    
#     volume_trade_ideal = IntegerField()  # 策略理想下单量
    volume_trade_simulate = IntegerField()  # 策略模拟盘下单量
    
    class Meta:
        database = db
        primary_key = CompositeKey('datetime_triger_tick', 'direction')
        
        
        
db.create_tables([SignalDb, TradeSimulateDb])
```

```python
SignalDb.create(strategy="temp", 
                symbol="temp", 
                signal_id="temp", 
                datetime_begin=datetime.datetime(2020, 1, 8, 10, 10, 50), 
                datetime_end=datetime.datetime(2020, 1, 8, 10, 10, 50), 
                direction="long", 
                price_limit_1=10, 
                price_limit_2=10, 
                price_stop=10, 
               volume_signal=10, duration_min=10, cancel_second=10,
               used_to_trade=False)
```

```python
TradeSimulateDb.create(strategy="temp", 
                symbol="temp", datetime_simulate_trade=datetime.datetime(2020, 1, 8, 10, 10, 50), 
                datetime_triger_tick=datetime.datetime(2020, 1, 8, 10, 10, 50),
                      direction="long",offset="open", price_trade_simulate=10, volume_trade_simulate=10
                      )
TradeSimulateDb.create(strategy="temp1", 
                symbol="temp1", datetime_simulate_trade=datetime.datetime(2020, 1, 10, 10, 10, 50), 
                datetime_triger_tick=datetime.datetime(2020, 1, 10, 10, 10, 50),
                      direction="long",offset="open", price_trade_simulate=10, volume_trade_simulate=10
                      )
TradeSimulateDb.create(strategy="temp1", 
                symbol="temp1", datetime_simulate_trade=datetime.datetime(2020, 1, 10, 10, 10, 50), 
                datetime_triger_tick=datetime.datetime(2020, 1, 10, 10, 10, 50),
                      direction="short",offset="open", price_trade_simulate=10, volume_trade_simulate=10
                      )
```

```python
q = TradeSimulateDb.update({TradeSimulateDb.datetime_simulate_trade: datetime.datetime(2020, 3, 10, 10, 10, 50)}).where((TradeSimulateDb.datetime_triger_tick == datetime.datetime(2020, 1, 10, 10, 10, 50)) & (TradeSimulateDb.direction == "long")) 
q.execute()
```
