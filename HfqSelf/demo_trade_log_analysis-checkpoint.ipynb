```python
import pandas as pd
import numpy as np
import os
from peewee import *
```

```python
def fetch_data(beging_datetime="2021-01-17", end_datetime="2021-01-18", symbol="rb2105", db_path="D:\Trader\.vntrader\database.db"):
    db = SqliteDatabase(db_path)
    market_log = pd.read_sql("select * from dbtickdata where (symbol='%s') and Date(datetime) between '%s' and '%s'" %(symbol, beging_datetime, end_datetime), db, parse_dates="datetime")
    return market_log
temp_df = fetch_data()
```

```python
temp_df.columns
```

```python
keep_columns = ["datetime", "last_price", "volume", "bid_price_1", "bid_volume_1", "ask_price_1", "ask_volume_1"]
df = temp_df[keep_columns]
df["datetime"] = df["datetime"]
```

```python
df.values.shape
```

```python
from prophet.predict import signal_output
```

```python
model = signal_output()
```

```python
import datetime
class SingleCombinedLog:
    def __init__(self, raw_signal, df_predict_interval):
        self.raw_signal = raw_signal
        self.df_predict_interval = df_predict_interval
dict_long = {}  # datetime: SingleCombinedLog
dict_short = {} 
for i, timestamp_beging in enumerate(df["datetime"]):
    if (i > 2000) & ((i % 20) == 0):
        train_df = df.iloc[i-2000:i, :]

        train_df["datetime"] = train_df["datetime"].apply(lambda x:x.timestamp())
        raw_signals = model.get_output_list(train_df.values)
        print(timestamp_beging)
        for j, signal in enumerate(raw_signals):
            timestamp_end = timestamp_beging + datetime.timedelta(minutes=signal["duration_min"])
            df_predict = df.loc[(df["datetime"] > timestamp_beging) & (df["datetime"] < timestamp_end)]
            if signal["direction"] == "long":
                dict_long[timestamp_beging] = SingleCombinedLog(signal, df_predict)
            elif signal["direction"] == "short":
                dict_short[timestamp_beging] = SingleCombinedLog(signal, df_predict)
```
